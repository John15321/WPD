```python
from functools import lru_cache
import time
import timeit

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from foo import (
    fibonacci,
    fibonacci_cache,
    fibonacci_lru_cache,
    foo_timer,
)
```

```python
n = 15

fib1 = [foo_timer(fibonacci, i) for i in range(1, n)]
fib2 = [foo_timer(fibonacci_cache, i) for i in range(1, n)]
fib3 = [
    foo_timer(fibonacci_lru_cache, i) for i in range(1, n)
]

nr = list(range(1, n))

plt.plot(nr, fib1, "-o", label="Native Fibonacci")
plt.plot(nr, fib2, "-o", label="Fibonacci memoization")
plt.plot(nr, fib3, "-o", label="Fibonacci @lru_cache")

plt.ylabel("Czas [s]")
plt.xlabel("n")
plt.xticks(range(0, n))
plt.title("Fibonacci(n) w zależności od czasu")
plt.grid(True)
plt.legend()
plt.savefig("fibs.jpg", dpi=500)
plt.show()
```

```python
df = pd.DataFrame(
    data={
        "n": list(range(1, n)),
        "fib1": fib1,
        "fib2": fib2,
        "fib3": fib3,
    }
)
df.to_csv("fibs.csv")
```
